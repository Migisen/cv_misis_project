## Метрика качества

В качестве метрики качества будут выбраны $MAE$ и $MSE$, так как они являются доминирующими метриками, используемыми для подсчета количественной ошибки предсказания числа людей. 

$$
MAE = \dfrac{\sum_{i=1}^n|y_i-x_i|}{n}; \\

MSE = \dfrac{1}{n}\sum_{i=1}^{n}(y_i - x_i)^2;
$$
где: \
$n$ - кол-во наблюдений \
$y_i$ - предсказанное моделью количество людей на фотографии \
$x_i$ - реальное еоличество людей на изображении \
